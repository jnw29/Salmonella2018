```python
import os
import pandas as pd
import numpy as np
import scipy.misc as smp
```

This block of code is for creating a color-coded representation fo the continents or geographical regions each strain comes from.

Colors:

1 - North America - Magenta

2 - South America - Yellow

3 - Europe - Blue

4 - Asia - Red

5 - Africa - Green

6 - Oceania - Cyan

7- Unknown / N/A - White

```python
#first, scrape the data and assign countries to continental regions

data = '/Users/jay.worley/Desktop/manuscript/Draft 3/2017-10-28_Seq_Info_Supplemental_modified.xlsx'
df = pd.read_excel(data, index_col=0)

#each country here: country_dict[country] = Continent # (see markdown)
country_dict = {}

#each seqID assigned to the country
SID_country_dict = {}

for name in df.index:
    country = df['Country'][name]
    SID_country_dict[name] = country
    if country not in list(country_dict.keys()):
        country_num = input(country+' Enter number for country:')
        country_dict[country] = country_num
```

```python
country_code_file = '2017-10-25_country_codes.tab'
file_text_lines = []
for country in country_dict:
    print(country, country_dict[country])
    file_text_lines.append(country+'\t'+str(country_dict[country]))
file_text = '\n'.join(file_text_lines)
with open(country_code_file, 'w') as f:
    f.write(file_text)
```

```python
##Tree order
with open('/Users/jay.worley/Desktop/RAxML Trees/RAxML_bipartitions.RAxML-concat-ccg_2-nd.nex', 'r') as f:
    nexus = f.read()
tree = nexus.split('begin trees;')[1].split('end;')[0].strip()
splits = ['S'+i.split(':')[0] for i in tree.split('S')[1:]]
for i in range(len(splits)):
    if ':' in splits[i]:
        splits[i] = splits[i].split(':')[0]
tree_order = splits
    
#color codes:

black = [0,0,0]
magenta = [255, 0 ,255]
yellow = [255, 255, 0]
blue = [0, 0, 255]
red = [255, 0, 0]
green = [0, 255, 0]
cyan = [0, 255, 255]
white = [255, 255, 255]
colors = [black, magenta, yellow, blue, red, green, cyan, white]

#create canvas
sample_height = 5
height = len(tree_order)*sample_height
width = 40
print(height, width)
data = np.zeros((height,width,3), np.uint8)

for i in range(len(tree_order)):
    country = SID_country_dict[tree_order[i]]
    color = colors[int(country_dict[country])]
    for h in range(i*sample_height, (i*sample_height)+sample_height):
        for w in range(0,width):
            data[h,w]=color

outfile = '2017-10-25_counrty-color-barcode.png'            
img = smp.toimage(data).save(outfile)
```

```python
#basic stats
country_count = []
continent_count = []

continents = ['index_adjustment', 'North America', 'South America', 'Europe', 'Asia', 'Africa', 'Oceania', 'Unknown']

for seq in tree_order:
    country = SID_country_dict[seq]
    continent = continents[int(country_dict[country])]
    country_count.append(country)
    continent_count.append(continent)
print(len(sorted(set(country_count))))
for country in sorted(set(country_count), key = lambda x: country_count.count(x)):
    print(country, country_count.count(country))
print('\n\n')
for continent in sorted(set(continent_count), key = lambda x: continent_count.count(x)):
    print(continent, continent_count.count(continent))
```
