```python
resultdir = './unzipped_results/' #redirect as needed
resultdirs = [i for i in os.listdir(resultdir) if os.path.isdir(resultdir+i)]
rd = {} #result dictionary rd[strain]={1:hit}
for result in resultdirs:
    rd[result]={}
    sf = os.path.join(resultdir,result+'/','summary.txt') #summary file
    with open(sf,'r') as f:
        st = f.read() #summary text
    foo = False
    matches = []
    keys = []
    for line in [i.strip() for i in st.split('\n\n\n\n')[1].split('\n')]:
        if line.startswith('REGION'):
            keys = [i for i in line.split(' ') if len(i) > 0]
            continue
            
        try:
            ls = [i for i in line.split(' ') if len(i) > 0]
            int(ls[0])
            match = [i for i in line.split(' ') if len(i) > 0]
            rd[result][int(ls[0])]=match
        except:
            continue
```
